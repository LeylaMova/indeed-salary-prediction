```python
import pandas as pd
import numpy as np
import tensorflow as tf
import tflearn
from tflearn.data_utils import to_categorical
from collections import Counter
from lib.preprocessing import clean_salary, clean_text, standardize
from lib.normalizing_city import map_city, suburban_vs_city

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_pickle('data/df.pkl')
```

```python
part_time = df[(df['salary'].str.contains('month') | 
                df['salary'].str.contains('week') | 
                df['salary'].str.contains('hour'))]
```

```python
df = df[~(df['salary'].str.contains('month') | 
          df['salary'].str.contains('week') | 
          df['salary'].str.contains('hour'))]
```

```python
df['salary'] = df['salary'].apply(clean_salary)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
df['company'] = df['company'].apply(clean_text)
df['summary'] = df['summary'].apply(clean_text)
df['title'] = df['title'].apply(clean_text)
df['city'] = df['city'].astype(str)
df['state'] = df['state'].astype(str)
df['title'] = df['title'].apply(standardize)
df['summary'] = df['summary'].apply(standardize)
df['city'] = df.apply(map_city, axis=1)
df['city'] = df['city'].apply(lambda x: x + ' ' + 'Urban' if 'Suburb' not in x else x)
```

```python
twentieth = df['salary'].quantile(0.20)
eightyfifth = df['salary'].quantile(0.85)
```

```python
ax = sns.distplot(df['salary'])
```

```python
def percentile_cls(row):
    
    salary = row['salary']    
    if salary < twentieth:
        return 0
    elif salary >= twentieth and salary < eightyfifth:
        return 1 
    else:
        return 2
```

```python
df['salary_class'] = df.apply(percentile_cls, axis=1)
```

```python
df.head(2)
```

```python
def combine_text(row):
    city = row['city'].lower()
    company = row['company']
    state = row['state'].lower()
    summary = row['summary']
    title = row['title']
    text = city + ' ' + company + ' ' + state + ' ' + summary + ' ' + title 
    
    return text
```

```python
df['text'] = df.apply(combine_text, axis=1)
```

```python
X = df['text'].values
y = df['salary_class'].values
```

```python
total_counts = Counter()
for text in X:
    for word in text.split(" "):
        total_counts[word] += 1

print("Total words in data set: ", len(total_counts))
```

```python
vocab = sorted(total_counts, key=total_counts.get, reverse=True)[:1000]
print(vocab[:30])
```

```python
print(vocab[-1], ': ', total_counts[vocab[-1]])
```

```python
word2idx = {}
for i, word in enumerate(vocab):
    word2idx[word] = i
```

```python
def text_to_vector(text):
    word_vec = np.zeros((1, len(vocab)))
    for word in text.split(" "):
        if word in word2idx.keys():
            word_vec[0][word2idx[word]] += 1
            
    return word_vec
```

```python
word_vectors = np.zeros((len(X), len(vocab)), dtype=np.int_)
for ii, text in enumerate(X):
    word_vectors[ii] = text_to_vector(text)
```

```python
# Printing out the first 5 word vectors
word_vectors[:5, :23]
```

```python
records = len(y)

shuffle = np.arange(records)
np.random.shuffle(shuffle)
test_fraction = 0.8

train_split, test_split = shuffle[:int(records*test_fraction)], shuffle[int(records*test_fraction):]
trainX, trainY = word_vectors[train_split,:], to_categorical(y[train_split], 3)
testX, testY = word_vectors[test_split,:], to_categorical(y[test_split], 3)
```

```python
# Network building
def build_model():
    # This resets all parameters and variables, leave this here
    tf.reset_default_graph()
    
    # Inputs
    net = tflearn.input_data([None, 1000])
    
    # Hidden layer(s)
    net = tflearn.fully_connected(net, 100, activation='ReLU')
    net = tflearn.fully_connected(net, 20, activation='ReLU')

    # Output layer(s)
    net = tflearn.fully_connected(net, 3, activation='softmax')
    net = tflearn.regression(net, optimizer='adam', 
                             learning_rate=0.1, 
                             loss='categorical_crossentropy')
    
    model = tflearn.DNN(net)
    
    return model
```

```python
model = build_model()
```

```python
# Training
model.fit(trainX, trainY, validation_set=0.2, show_metric=True, batch_size=128, n_epoch=10)
```

```python
model.fit(trainX, trainY, validation_set=0.1, show_metric=True, batch_size=128, n_epoch=100)
```

```python
model.fit(trainX, trainY, validation_set=0.1, show_metric=True, batch_size=128, n_epoch=200)
```

```python
model.fit(trainX, trainY, validation_set=0.1, show_metric=True, batch_size=128, n_epoch=300)
```

```python
predictions = (np.array(model.predict(testX))[:,0] >= 0.5).astype(np.int_)
test_accuracy = np.mean(predictions == testY[:,0], axis=0)
print("Test accuracy: ", test_accuracy)
```
